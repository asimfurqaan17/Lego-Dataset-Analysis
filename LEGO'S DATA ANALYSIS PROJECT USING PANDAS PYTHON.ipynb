#### MERGING THE DATA SETS

```python
merged = df.merge(theme,left_on='parent_theme', right_on='name')
merged.head()
```

#### DROPPING THE DUPLICATE NAME

```python
merged = df.merge(theme,left_on='parent_theme', right_on='name')
merged.drop(columns='name_y', inplace = True)
merged.head()
```

```python
licensed=merged[merged['is_licensed'] ==True]
licensed.head()
```

#### CLEANING THE DATA OF NaN's

```python
merged[merged['set_num'].isnull()].shape  #This gives us the number of NaN's rows in "star_wars"
```

```python
licensed.head() #We have to drop the NaN's from this
```

```python
licensed=licensed.dropna(subset='set_num')
licensed.shape[0]
```

#### ANSWERING THE FIRST QUESTION

```python
star_wars=licensed[licensed['parent_theme']=='Star Wars']
star_wars.shape[0]
```

```python
the_force=int(star_wars.shape[0]/licensed.shape[0]*100)  #Answer to Question Number 1
print(the_force)
```

#### ADDING A COUNT ROW

```python
licensed_sorted=licensed.sort_values('year')
licensed_sorted['count'] = 1
licensed_sorted.head()
```

```python
summed_df=licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()
summed_df.head()
```

```python
max_df=summed_df.sort_values('count',ascending=False).drop_duplicates(['year'])
max_df.head()
```
